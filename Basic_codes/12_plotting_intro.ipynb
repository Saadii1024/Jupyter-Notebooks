# Plotting Stuff 

```python
nums = list(range(1, 13))

power_2 = []
power_3 = []

for i in nums:
    power_2.append(pow(2, i))
    power_3.append(pow(3, i))

print(nums)
print(power_2)
print(power_3)
```

```python
import matplotlib.pyplot as plt
```

```python
# only needed in jupyter notebook 
%matplotlib inline  
```

```python
plt.figure()
plt.plot(nums, power_2)
```

```python
plt.figure()
plt.plot(nums, power_2)
plt.plot(nums, power_3)
```

```python
plt.figure() 
plt.plot(nums, power_2, label='2^nums')
plt.plot(nums, power_3, label='3^nums')
plt.xlabel('Numbers')
plt.ylabel('2^nums & 3^nums')
plt.legend()
```

## Case study: Time taken by fib(n)

```python
# our old fib function 
def fibslow(n): 
    if n == 0 or n == 1 :
        return n 
    
    else: 
        return fibslow(n-1) + fibslow(n-2)
    
def fibfast(n): 
    if n <= 1: return n 
 
    a, b = 0, 1 
    
    for i in range(1, 8):
        a, b = b, a + b
        
    return a
```

```python
from_n, to_n = 1, 30

X = list(range(from_n, to_n))
print(X)
```

```python
def time_function(fn, start, end): 
    from datetime import datetime
    
    times = [] 
    for i in range(start, end): 
        start_time = datetime.now() 
        
        fn(i)  # actual call to the function, don't care about value 
        
        end_time = datetime.now() 
        time_taken = end_time - start_time   # returns a timedelta 
        time_taken = time_taken.microseconds
        
        times.append(time_taken)
    
    return times        
```

```python
fibslow_times = time_function(fibslow, from_n, to_n)
fibfast_times = time_function(fibfast, from_n, to_n)

print(fibslow_times) 
print(fibfast_times)
```

```python
plt.figure() 
plt.plot(X, fibslow_times, label='fibslow')
plt.plot(X, fibfast_times, label='fibfast')
plt.xlabel('n')
plt.ylabel('time (microseconds)')
plt.legend()
```
